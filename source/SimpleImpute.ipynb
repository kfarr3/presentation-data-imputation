```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [10, 10]
pd.options.display.max_columns = None
```

```python
num_samples = 10
np.random.seed=33
df_orig = pd.Series(np.random.randint(0, 10, num_samples).astype(float))
```

```python
df_orig.plot()
plt.title("Original Data", fontsize=30)
plt.savefig('original.png')
```

```python
df_missing = df_orig.copy()
df_missing[4]=np.nan
df_missing.plot()
plt.title("Missing Data", fontsize=30)
plt.savefig('missing.png')
```

```python
df_missing.fillna(method='ffill').plot(color='r')
df_orig.plot()
plt.title("Forward Fill", fontsize=30)
plt.savefig("forward.png")
```

```python
df_missing.fillna(method='bfill').plot(color='r')
df_orig.plot()
plt.title("Back Fill", fontsize=30)
plt.savefig("back.png")
```

```python
df_missing.interpolate().plot(color='r')
df_orig.plot()
plt.title("Interpolate", fontsize=30)
plt.savefig("interpolate.png")
```

```python
df_missing.fillna(df_missing.mean()).plot(color='r')
df_orig.plot()
plt.title("Mean", fontsize=30)
plt.savefig("mean.png")
```
